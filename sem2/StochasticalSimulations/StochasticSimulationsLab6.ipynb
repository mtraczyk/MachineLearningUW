```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
# generate n poisson numbers
n = 1000000
p_lam = 100.
poisson = np.random.exponential(1 / p_lam, n)

b_p = 0.7
bernoulli = np.random.binomial(1, b_p, n)

poisson = np.cumsum(poisson)
poisson = poisson[bernoulli == 1]

diffs = np.diff(poisson)
print(np.mean(diffs))

plt.hist(diffs, bins=20)
plt.show()
```

```python
p_lam_1 = 5.
p_lam_2 = 3.
n = 1000000

poisson_1 = np.cumsum(np.random.exponential(1 / p_lam_1, n))
poisson_2 = np.cumsum(np.random.exponential(1 / p_lam_2, n))

poisson = np.concatenate((poisson_1, poisson_2))
diffs = np.diff(np.sort(poisson))
diffs = diffs[:n]

print(np.mean(diffs))

plt.hist(diffs, bins=20)
plt.show()
```

```python
# Markov Chains
# X_{n+1} = aX_{n} + eps
n = 2000
a = 1.
base = 0
random_walk = np.zeros(n)
random_walk[0] = base

for i in range(1, n):
  random_walk[i] = a * random_walk[i - 1] + np.random.randn(1)[0]

plt.plot(random_walk)
plt.show()
```
